# Урок 4. Непрерывная случайная величина

1. Случайная непрерывная величина A имеет равномерное распределение на промежутке (200, 800].
Найдите ее среднее значение и дисперсию.

```python
# M(A) = (a+b)/2 D(A)=(b-a)^2/12
M = (200+800)/2
print(f'Среднее значение M(A)= {M}')
D = (800-200)**2/12
print(f'Дисперсия D(A)= {D}')
```

2. О случайной непрерывной равномерно распределенной величине B известно, что ее дисперсия равна 0.2.
Можно ли найти правую границу величины B и ее среднее значение зная, что левая граница равна 0.5?
Если да, найдите ее.

```python
import math
# D(A)=(b-a)^2/12
# sqrt(12*D(a))=b-a
# b = sqrt(12*D(a))+a
Da = 0.2
a = 0.5
b = math.sqrt(12*Da)+a
# M(A) = (a+b)/2
Ma = (a+b)/2
print(f'Правая граница {b}\n Среднее значение M(a) = {Ma}')
```

3. Непрерывная случайная величина X распределена нормально и задана плотностью распределения f(x) = (1 / (4 * sqrt(2pi))) * exp((-(x+2)**2) / 32)

Найдите:

а). M(X)

б). D(X)

в). std(X) (среднее квадратичное отклонение)

> Решение:

По формуле Гуасса
a = M(x) б^2 = D(x)

Из формулы нашей функции М(x) = 4,

2*б^2=32 => b^2 => D(x) = 16

std(X)= sqrt(Dx) = 4

> Ответ:

М(x) = 4

D(x) = 16

std(X) = 4



4. Рост взрослого населения города X имеет нормальное распределение.
Причем, средний рост равен 174 см, а среднее квадратичное отклонение равно 8 см.

Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:
а). больше 182 см
б). больше 190 см
в). от 166 см до 190 см
г). от 166 см до 182 см
д). от 158 см до 190 см
е). не выше 150 см или не ниже 190 см
ё). не выше 150 см или не ниже 198 см
ж). ниже 166 см.

```python
m = 174
x1 = 182
x2 = 190
sdt = 8
z1 = (x1-m)/sdt
print(f'Z182({z1}) = 0,8413') 
print(f'a) Вероятность что человек выше 182см = 0,1587') 
z2 = (x2-m)/sdt
print(f'Z190({z2}) = 0,9772') 
print(f'б) Вероятность что человек выше 190см = 0,0228') 
z166 = (166 - m)/sdt
print(f'Z166({z166}) = 0.1587')
print(f'в) Вероятность что человек от 166 до 190 см = {0.9772 - 0.1587}') 
print(f'г) Вероятность что человек от 166 до 182 см = {0.8413 - 0.1587}') 
z158 = (158 - m)/sdt
print(f'Z158({z158}) = 0.0228')
print(f'д) Вероятность что человек от 158 см до 190 см = {0.9772 - 0.0228}') 
z150 = (150 - m)/sdt
print(f'Z150({z150}) = 0.0013')
print(f'е) Вероятность что человек не выше 150 см или не ниже 190 см = {0.0013 + 0.0228}')
z198 = (198 - m)/sdt
print(f'Z198({z198}) = 0.9987')
print(f'ё) Вероятность что человек не выше 150 см или не ниже 198 см = {0.0013 + (1-0.9987)}') 
print(f'д) Вероятность что человек ниже 166 см. = 0.1587')
```
