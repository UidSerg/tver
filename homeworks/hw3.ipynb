1. Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое, среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.

```python
import numpy as np
arr = np.array([100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150])
arr.sort()
print(f'упорядоченный массив {arr}')
n = len(arr)
print(f'Длинна массива {n} элементов')
sa = np.sum(arr)/len(arr)
print(f'среднее арифметическое = {sa}')
## S^2 = (Sum (arr[i]-sA)^2)/n-1
sum = 0
for i in range(n):
    sum=sum+(arr[i]-sa)**2
Skv=sum/(n-1)
print(f'среднее квадратичное отклонение выборки = {Skv}')
print(f'смещенная оценка')
print(f'стандартное отклонение {np.std(arr)}')
print(f'дисперсия {np.var(arr)}')
print(f'корень {np.sqrt(np.var(arr))}')
print(f'не смещенная оценка')
print(f'стандартное отклонение {np.std(arr, ddof=1)}')
print(f'дисперсия {np.var(arr, ddof=1)}')
print(f'корень {np.sqrt(np.var(arr, ddof=1))}')
```

2. В первом ящике находится 8 мячей, из которых 5 - белые. Во втором ящике - 12 мячей, из которых 5 белых. Из первого ящика вытаскивают случайным образом два мяча, из второго - 4. Какова вероятность того, что 3 мяча белые?

```python
from math import factorial
# хотим проверить что событие произойдет k раз из n
def comb(n, k):
    return factorial(n)/(factorial(k)*factorial(n-k))

# 2 белых из к1+1 белый из к2 или 1б из к1 и 2б из к2 или 0б из к1 и 3б из к2
x=comb(5,2)*comb(3,0)/comb(8,2)*comb(5,1)*comb(7,3)/comb(12,4)+comb(5,1)/comb(8,2)*comb(5,2)/comb(12,4)+comb(5,0)/comb(8,2)*comb(5,3)/comb(12,4)
print(f'вероятность того, что 3 мяча белые = {round(x,4)}')
```

4. В университет на факультеты A и B поступило равное количество студентов, а на факультет C студентов поступило столько же, сколько на A и B вместе. Вероятность того, что студент факультета A сдаст первую сессию, равна 0.8. Для студента факультета B эта вероятность равна 0.7, а для студента факультета C - 0.9. Студент сдал первую сессию. Какова вероятность, что он учится: a). на факультете A б). на факультете B в). на факультете C?


```python
#  доля студентов pa=1/4, pb=1/4, pc=2/4
pa = 0.8
pb = 0.7
pc = 0.9
# вероятность события
P = 1/4*pa+1/4*pb+2/4*pc
va = 1/4*pa/P
vb = 1/4*pb/P
vc = 2/4*pc/P
print(f'вероятность что студент учится\n в A: {round(va,4)},\n в B: {round(vb,4)},\n в C: {round(vc,4)}')
```

5. Устройство состоит из трех деталей. Для первой детали вероятность выйти из строя в первый месяц равна 0.1, для второй - 0.2, для третьей - 0.25. Какова вероятность того, что в первый месяц выйдут из строя: а). все детали б). только две детали в). хотя бы одна деталь г). от одной до двух деталей?

```python
p1 = 0.1
p2 = 0.2
p3 = 0.25
P = p1*p2*p3
print(f'вероятность выхода из строя всех деталей = {round(P,4)}')
# только две 1и2 или 1и3 или 2и3
Pt =p1*p2*(1-p3)+p1*p3*(1-p2)+p2*p3*(1-p1)
print(f'вероятность выхода любых двух деталей = {round(Pt,4)}')
# хотябы одна 1 или 2 или 3 или (1и2 или 1и3 или 2и3) или все
Po=P+Pt+p1*(1-p2)*(1-p3)+p2*(1-p1)*(1-p3)+p3*(1-p1)*(1-p2)
print(f'вероятность выхода из строя хотябы одной детали {round(Po,4)}')
print(f'вероятность выхода от одной до двух {round(Pt+Po,4)}')
```
